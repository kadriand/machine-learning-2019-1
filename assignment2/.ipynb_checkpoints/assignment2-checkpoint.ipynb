```python
import numpy as np
from sklearn import datasets
from sklearn import metrics
import scipy.stats as stats
import matplotlib.pyplot as plt
import random


iris = datasets.load_iris()
y = iris.target
means = np.zeros((3, 2))
covariances = np.zeros((3, 2, 2))
iris_data = iris.data[:, [1, 3]]  # we take the second and fourth
classes_list = np.unique(iris.target)


def plot_classes(points, classes, title=None, lines=[]):
    x_min, x_max = points[:, 0].min() - .5, points[:, 0].max() + .5
    y_min, y_max = points[:, 1].min() - .5, points[:, 1].max() + .5
    plt.figure(3, figsize=(8, 6))
    plt.clf()
    # Plot the training points
    fig, ax = plt.subplots()
    ax.scatter(points[:, 0], points[:, 1], c=classes, cmap=plt.cm.Set1, edgecolor='k')
    axis_labels = [i for j, i in enumerate(iris.feature_names) if j in [1, 3]]
    plt.xlabel(axis_labels[0])
    plt.ylabel(axis_labels[1])
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    ax.grid(True)
    if len(lines) > 0:
        for serie in lines:
            plt.plot(serie[:, 0], serie[:, 1], '-o')
    if title:
        plt.title(title)
    plt.show()
```

## 1.A
For the estimation of &#963;, the dataset was separated in three subsets, one for each class. To each set the mean was computed and those mean values were subtracted from the points of each set. The purpose was to obtain a new data set with mean value of [0,0]. Finally the variance for each column was computed and the mean of the two variances became into &#963;. A plot of random points generated using the distribution probabilites is shown.

```python
def sigma_a():
    class_data = [0] * 3
    iris_mean_0 = np.zeros((len(iris_data), 3))
    for class_number in classes_list:
        class_data[class_number] = iris_data[np.where(iris.target == class_number), :][0]
        means[class_number] = [class_data[class_number][:, 0].mean(), class_data[class_number][:, 1].mean()]
        for row in np.where(iris.target == class_number)[0]:
            iris_mean_0[row] = [iris_data[row][0] - means[class_number][0], iris_data[row][1] - means[class_number][1], class_number]
    sigma_found = np.mean([np.var(iris_mean_0[:, 0]), np.var(iris_mean_0[:, 1])])
    return sigma_found


def plot_1_a(covariance_matrix):
    # PLOTTING RESULTS
    dist_size = 600
    gauss = np.random.multivariate_normal((0, 0), covariance_matrix, dist_size)
    gauss_classes = [0] * dist_size
    for i, element in enumerate(gauss):
        element_class = i % 3
        gauss_classes[i] = element_class
        gauss[i] = [element[0] + means[element_class][0], element[1] + means[element_class][1]]
    plot_classes(gauss, gauss_classes)


sigma = sigma_a()
diagonal_covariance = np.identity(2) * sigma
print(f"Sigma :\n {sigma}")
print(f"Covariances Matrix :\n {diagonal_covariance}")
plot_1_a(diagonal_covariance)
```

## 1.B
Bayesian discriminant functions were defined using the previously obtained Bivariate Gaussian distributions as the probability function for each class. The covariances matrix used in all functions was the one based on &#963;I. The classifier_b function summarizes the discriminants

```python
def classifier_1_b(point, covariance_matrix):
    class_prob = np.zeros(3)
    for class_val in classes_list:
        class_prob[class_val] = stats.multivariate_normal.pdf(point, means[class_val], covariance_matrix)
    return class_prob.argmax()


diagonal_covariance = np.identity(2) * sigma_a()
a = [3.2, .2]
b = [2.7, 1.5]
c = [3, 2.05]

print("Test points (near to the means of each class)\n")
print(f"{a} : {classifier_1_b(a, diagonal_covariance)}")
print(f"{b} : {classifier_1_b(b, diagonal_covariance)}")
print(f"{c} : {classifier_1_b(c, diagonal_covariance)}")
```

## 1.D
The lines shown in the figure were obtain of this way:

A linear discriminant was obtained according to the formulas *5.23*:

$$g_i(x)=w_i^Tx+w_{i0}$$

Once the equation for each class in the discriminant were obtained, the equations for the lines delimiting  the class regions were deducted having in mind that for the points of those lines the discriminant must classify them in both regions. In other words, the line separating the class __k__ and **l** is the interception between their corresponding equations in the discriminant. 
$$g_k(x)=g_l(x)$$

For a class with 2D points

$$w_k^Tx+w_{k0}=w_l^Tx+w_{l0}$$

$$(w_k^T-w_l^T)x=w_{l0}-w_{k0}$$

$$(w_k^T-w_l^T)[x_1,x_2]=w_{l0}-w_{k0}$$

Once $x_2$ (or $x_1$) is isolated in the last equation, the result is the equation of the line dividing the classes **k** and __l__.

The result of applying this strategy are 3 equations (separating the classes 0-1, 0-2 and 1-2). At the end the equation whose intercept with 0 was neither the highest nor the lowest was discarted.

The _line parameters_ printed in the result corresponds to:
$$line \; parameters=[m,y_0]_k$$
Line equation
$$y=m x+y_0$$

```python
def parameters_1_d(prior_probs=None):
    diagonal_covariance = np.identity(2) * sigma_a()
    common_cov_s = np.zeros((2, 2))
    set_size = len(iris_data)
    class_size = np.zeros(3)
    for class_number in classes_list:
        class_size[class_number] = len(np.where(iris.target == class_number)[0])
        common_cov_s = common_cov_s + diagonal_covariance * class_size[class_number] / set_size

    prior_probs = prior_probs if prior_probs is not None else class_size / set_size
    s_inverted = np.linalg.inv(common_cov_s)
    wi = [np.matmul(s_inverted, means[class_number]) for class_number in classes_list]
    wi0 = [-1 / 2 * np.matmul(np.matmul(np.transpose(means[class_number]), s_inverted), means[class_number]) + np.log(prior_probs[class_number]) for class_number in classes_list]
    return wi, wi0


def classifier_d(point, wi, wi0):
    class_prob = np.zeros(3)
    for class_val in classes_list:
        class_prob[class_val] = np.matmul(wi[class_val], point) + wi0[class_val]
    return class_prob.argmax()


def lines_params_d(prior_probs=None):
    (wi, wi0) = parameters_1_d(prior_probs)

    dw0_10 = wi0[1] - wi0[0]
    dw0_20 = wi0[2] - wi0[0]
    dw0_21 = wi0[2] - wi0[1]
    dw_10 = [wi[1][0] - wi[0][0], wi[1][1] - wi[0][1]]
    dw_20 = [wi[2][0] - wi[0][0], wi[2][1] - wi[0][1]]
    dw_21 = [wi[2][0] - wi[1][0], wi[2][1] - wi[1][1]]

    lines_params = np.array([
        [-dw_10[0] / dw_10[1], -dw0_10 / dw_10[1]],
        [-dw_20[0] / dw_20[1], -dw0_20 / dw_20[1]],
        [-dw_21[0] / dw_21[1], -dw0_21 / dw_21[1]]
    ])

    return [lines_params[np.argwhere(lines_params[:, 1] == lines_params[:, 1].max())[0]][0], lines_params[np.argwhere(lines_params[:, 1] == lines_params[:, 1].min())[0]][0]]


(wi, wi0) = parameters_1_d()
print(f"wi : \n {wi}")
print(f"wi0 : \n {wi0}")
lines_params = lines_params_d()
print(f"\nlines parameters mX+Y0 vals: [m, Y0]_1...: \n{lines_params} ")
lines = np.zeros((len(lines_params), 2, 2))
for (idx, line_params) in enumerate(lines_params):
    lines[idx][0] = [0, line_params[1]]
    lines[idx][1] = [5, line_params[0] * 5 + line_params[1]]
plot_classes(iris_data, iris.target, lines=lines)


```

It can be seen how the lines can easily separarate the regions shown in __1.C__

## 1.E
Three Prior Probabilites are displayed:
$$P(C_k)=P(C_k) (the\;original)$$

$$P(C_k)=[0.3, 0.3, 0.4]$$

$$P(C_k)=[0.05, 0.45, 0.5]$$

```python
def class_boundaries_e(prior_probs=None):
    lines_params = lines_params_d(prior_probs)
    print(f"prior probabilities : \n{prior_probs} ")
    print(f"\nlines parameters mX+Y0 vals: [m, Y0]_1...: \n{lines_params} ")

    lines = np.zeros((len(lines_params), 2, 2))
    for (idx, line_params) in enumerate(lines_params):
        lines[idx][0] = [0, line_params[1]]
        lines[idx][1] = [5, line_params[0] * 5 + line_params[1]]
    plot_classes(iris_data, iris.target, lines=lines, title=f"PC = {prior_probs if prior_probs else ''}")


class_boundaries_e()
class_boundaries_e([.3, .3, .4])
class_boundaries_e([.05, .45, .5])
```

As the figures shows the differences of the dividing lines are very slight when varying the Prior Probabilities. To study the reasons, in one of the formulas 5.23 can be seen the Prior Probabilities term:

$$w_{i0}=-1/2 m_i^T+log \hat{P}(C_i)$$

An as can be noted in the _lines parameters_ output ${[[m,y_0]_1,...]}$, for the equations of the new lines, the modification of thr Prior probabilites only affects the $y_0$ (interception with _y_ axis) parameter of the lines (it displaces vertically), the slope of the lines remains unaffected.

## 2.A
Using the formula 3.14 and the common covariance matrix for all classes computed in __1.A__

```python
def classifier_2(point, loss, covariance_matrix):
    class_prob = np.zeros(4)
    for class_val in classes_list:
        class_prob[class_val] = stats.multivariate_normal.pdf(point, means[class_val], covariance_matrix)
    class_prob[3] = 1 - loss
    return class_prob.argmax()
```

```python
diagonal_covariance = np.identity(2) * sigma_a()
a = [3.2, .2]
b = [2.7, 1.5]
c = [3, 2.05]
loss_lambda = 0.5
print(f"{a} : {classifier_2(a, loss_lambda, diagonal_covariance)}")
print(f"{b} : {classifier_2(b, loss_lambda, diagonal_covariance)}")
print(f"{c} : {classifier_2(c, loss_lambda, diagonal_covariance)}")
```

## 3.A
Usin random numbers to build the covariance matrixes, but keepin the off-diagonals with the same value

```python
def make_random_covariances_3(data_set=iris_data, data_clases=iris.target):
    class_data = [0] * 3
    for class_number in classes_list:
        diagonal_val = np.random.uniform()
        covariances[class_number] = [[np.random.uniform(), diagonal_val], [diagonal_val, np.random.uniform()]]
        class_data[class_number] = data_set[np.where(data_clases == class_number), :][0]
        means[class_number] = [class_data[class_number][:, 0].mean(), class_data[class_number][:, 1].mean()]
        print(f"Random covariance matrix for {iris.feature_names[class_number]}: \n {covariances[class_number]}")
        print(f"Mean for {iris.feature_names[class_number]} :\n {means[class_number]}")
        

def plot_3_a():
    # PLOTTING RESULTS
    samples_per_class = 200
    dist_points = np.zeros((samples_per_class * len(classes_list), 2))
    dist_classes = np.zeros(samples_per_class * len(classes_list))
    for class_number in classes_list:
        gauss = np.random.multivariate_normal(means[class_number], covariances[class_number], samples_per_class)
        for i, point in enumerate(gauss):
            idx = i + class_number * samples_per_class
            dist_classes[idx] = class_number
            dist_points[idx] = point
    plot_classes(dist_points, dist_classes)
    
    
make_random_covariances_3()
plot_3_a()
```

## 3.B
Usin random numbers to build the covariance matrixes, but keepin the off-diagonals with the same value

```python
def make_covariances_3(data_set=iris_data, data_clases=iris.target):
    class_data = [0] * 3
    for class_number in classes_list:
        class_data[class_number] = data_set[np.where(data_clases == class_number), :][0]
        covariances[class_number] = np.cov(class_data[class_number][:, 0], class_data[class_number][:, 1], rowvar=False)
        means[class_number] = [class_data[class_number][:, 0].mean(), class_data[class_number][:, 1].mean()]
        print(f"Random covariance matrix for '{iris.feature_names[class_number]}' class: \n {covariances[class_number]}")
        print(f"Mean for '{iris.feature_names[class_number]}' class :\n {means[class_number]}")

        
iris_idxs = [i for i in range(len(iris_data))]
random.shuffle(iris_idxs)
samples_size = int(len(iris_data) * 0.8)
iris_training = iris_data[iris_idxs[0:samples_size]]
iris_training_classes = iris.target[iris_idxs[0:samples_size]]
make_covariances_3(iris_training, iris_training_classes)
plot_classes(iris_training, iris_training_classes, "Training set 80%")
```

```python
def iris_confusion_matrix(real, predicted, title="Confusion matrix"):
    classes_list = np.unique(iris.target)
    confusion_matrix = metrics.confusion_matrix(real, predicted, classes_list)
    print(f"confusion_matrix:\n {confusion_matrix}")
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(confusion_matrix)
    plt.title(title)
    fig.colorbar(cax)
    ax.set_xticklabels([''] + iris.feature_names)
    ax.set_yticklabels([''] + iris.feature_names)
    plt.xlabel('Predicted')
    plt.ylabel('Real')
    plt.show()

    
def classifier_3_c(point):
    class_prob = np.zeros(3)
    for class_val in classes_list:
        class_prob[class_val] = stats.multivariate_normal.pdf(point, means[class_val], covariances[class_val])
    return class_prob.argmax()


iris_test = iris_data[iris_idxs[samples_size - 1:len(iris_data)]]
iris_test_classes = [classifier_3_c(point) for point in iris_test]
plot_classes(iris_test, iris_test_classes, "Testing set 20%")
iris_test_real_classes = iris.target[iris_idxs[samples_size - 1:len(iris_data)]]
iris_confusion_matrix(iris_test_real_classes, np.asarray(iris_test_classes))
```
